# Panorama do COVID-19 no Brasil

COVID-19 é uma doença infecciosa causada pelo recém-descoberto coronavírus.

Transmitido principalmente por meio de gotículas provenientes de tosses ou espirros de pessoas infectadas, a gravidade dos sintomas varia muito de pessoa para pessoa.

Fato é, não se sabe muita coisa a respeito do COVID-19. Estudos estão sendo realizados no mundo todo, porém os resultados ainda não são conclusivos e definitivos.


<center><img alt="Colaboratory logo" width="70%" src="https://github.com/carvalhopinheiro/Coisas-Salvas/blob/main/covidult.png?raw=true"></center>

Até o presente momento, observa-se que cerca de 80% dos casos confirmados são assintomáticos e rápidos. A maioria das pessoas que se encaixam nesse grupo, se recupera sem nenhuma sequela.

No entanto, 15% das pessoas terão infecções graves e precisarão de oxigênio. O restante das pessoas, que representam 5%, serão classificadas como infecções muito graves e precisarão de ventilação assistida, por meio de respiradores mecânicos em ambiente hospitalar.

Com o objetivo de elevar a consciência situacional a respeito do COVID-19 no Brasil, irei realizar uma análise sobre os dados públicos da doença.

## Obtenção dos Dados
Para realizar a obtenção dos dados vamos importar as bibliotecas necessarias para a analise de dados e geração de graficos.

Importaremos as biliotecas mais famosas para essa área de análise e visualização de dados

* Pandas
* Matplotlib


Os dados utilizados neste projeto são oriundos do repositório do [Our World in Data](https://github.com/owid/covid-19-data/tree/master/public/data), portal que tem como objetivo deixar o conhecimento sobre os grandes problemas do mundo de uma forma acessível e compreensível, sob o lema "Pesquisa e dados para progredir contra os maiores problemas do mundo". O conjunto fornecido abrange dados de casos confirmados e mortes provocadas pela doença, de hospitalizações, internações, testes, vacinação e entre outras variáveis.

O grande desafio de analisar uma nova doença é a imprevisibilidade do vírus em si, pois ele pode gerar novas variantes e exigir novos estudos, como são os casos da variante Ômicron e Delta. Outro desafio é o contexto a qual o vírus está inserido, se é um país sub-desenvolvido ou não, se as autoridades tem adotado as medidas corretas no combate a doença, etc. Além disso, há o fator da cultura de dados no país em estudo. Apesar de ser transparente em relação ao seus dados, havendo o Portal da Transparência, o próprio dados.gov, o Brasil não vai bem nesse sentido, pois a qualidade dos dados fornecidos não é tão boa, podendo haver conflito entre os dados do governo federal e os dados estaduais, o que pode dificultar e até inviabilizar alguns estudos.

#### Dicionário de Variáveis
Vou informar algumas das variáveis das quais acho essencial o entendimento, para contextualização de todas as variaveis recomendo entrar no [Site](https://github.com/owid/covid-19-data/tree/master/public/data)  na qual os dados foram retirados.




*  ``iso_code``	- ISO 3166-1 alfa-3 - códigos de três letras do país 
*  ``continent``	- Continente da localização geográfica
*  ``location``	- Localização geográfica
*  ``date``	- Data de observação
*  ``median_age``	- Idade média da população, projeção da ONU para 2020
*   ``extreme_poverty``	- Parcela da população que vive em pobreza extrema, ano mais recente disponível desde 2010
*  ``hospital_beds_per_thousand``	- Leitos hospitalares por 1.000 pessoas, ano mais recente disponível desde 2010
*  ``population_density``	- Número de pessoas dividido por área terrestre, medido em quilômetros quadrados, ano mais recente disponível
*   ``population``	- População
*   ``total_vaccinations``	- Número total de COVID-19 vacinação doses administradas
*   ``people_vaccinated``	- Número total de pessoas que receberam pelo menos uma dose da vacina
*  ``people_fully_vaccinated``	- Número total de pessoas que receberam todas as doses prescritas pelo protocolo de vacinação
*   ``icu_patients``	- Número de COVID-19 pacientes em unidades de cuidados intensivos (UTIs) num dado dia
*   ``hosp_patients`` -	Número de pacientes COVID-19 no hospital em um determinado dia
*   ``total_deaths``	- Total de mortes atribuídas a COVID-19
*   ``new_deaths`` - Novas mortes atribuídas a COVID-19
*  ``total_cases`` - Total de casos confirmados de COVID-19
*  ``new_cases`` -	Novos casos confirmados de COVID-19
*  ``stringency_index`` - 	Governo Response Índice Rigor: medida composta com base em 9 indicadores de resposta incluindo o fechamento de escolas, fechamento de locais de trabalho, e proibição de viajar, escalonados para um valor de 0 a 100 (100 = resposta mais estrita)
* ``aged_70_older`` - 	Parcela da população com 70 anos ou mais em 2015

* ``excess_mortality_cumulative``	Diferença percentual entre o número cumulativo de mortes desde 1 de janeiro de 2020 e as mortes projetadas cumulativas para o mesmo período com base nos anos anteriores. Para obter mais informações
* ``excess_mortality_cumulative_absolute``	Diferença cumulativa entre o número relatado de mortes desde 1 de janeiro de 2020 e o número projetado de mortes para o mesmo período com base nos anos anteriores. Para obter mais informações
* ``excess_mortality_cumulative_per_million``	Diferença cumulativa entre o número de mortes relatado desde 1º de janeiro de 2020 e o número projetado de mortes para o mesmo período com base nos anos anteriores, por milhão de pessoas. Para obter mais informações
*  ``total_boosters	número`` total de doses de reforço COVID-19 administrado de vacinação (doses administradas para além do número prescrito pelo protocolo de vacinação)
* `` total_boosters_per_hundred``	Número total de COVID-vacinação de reforço 19 doses administradas por 100 pessoas na população total
* ``icu_patients``	Número de COVID-19 pacientes em unidades de cuidados intensivos (UTIs) num dado dia
* ``icu_patients_per_million``	Número de COVID-19 pacientes em unidades de terapia intensiva (UTIs) em um determinado dia por 1.000.000 de pessoas
* ``hosp_patients``	Número de pacientes COVID-19 no hospital em um determinado dia
